```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import Data as data
```

```python
def sector_data_fetch(sector_ticker, start_date,end_date):
    df_sector = pd.DataFrame()
    for i in sector_ticker:
        df = data.get_ETF_data(i, start_date, end_date)
        df_sector[i] = df['adj_close']
    return df_sector
```

# Airline Sector Data

```python
airline_tikers = ['UAL','AAL','ALK','JBLU','DAL','LUV','SKYW','RYAAY','CPA',
                  'ALGT','SAVE','SNCY','BA','NCLH','VLRS','MESA','HA','AZUL',
                  'ATSG','GOL']
```

```python
df_airline = sector_data_fetch(airline_tikers, '2021-03-17', '2024-01-01')
df_airline = df_airline.loc[:,df_airline.notna().all()]
df_airline
```

# Utility Sector Data

```python
utility_tikers = ['DUK','SO','NEE','D','SRE','EXC','CEG','PCG','AEP','GEV',
                  'XLU','RSPU','FUTY','IDU','VPU','PEG','XEL','NGG','WEC','AEE',
                  'ES','AWK','ETR','EIX','DTE','FE']

df_utility = sector_data_fetch(utility_tikers, '2021-03-17', '2024-01-01')
df_utility = df_utility.loc[:,df_utility.notna().all()]
df_utility
```

```python
df_utility = df_utility.loc[:,df_utility.notna().all()]
```

# Technology Sector

```python
tech_tikers = ['NVDA','AMD','AAPL','MSFT','CRM','CRWD','AVGO','INTC','ORCL','QCOM',
               'MU','INTU','AMAT','TTD','LRCX','NXPI','GOOG','ADBE','CSCO','TXN',
               'META','SNOW','DELL']

df_tech = sector_data_fetch(tech_tikers, '2021-03-17', '2024-01-01')
df_tech = df_tech.loc[:,df_tech.notna().all()]
df_tech
```

# Data Concatenation

```python
df_concat = pd.concat([df_airline.iloc[:,0:20],df_utility.iloc[:,0:20],df_tech.iloc[:,0:20]],axis=1).T.reset_index(names='ticker',)
label = np.concatenate((np.ones(20,dtype=int), np.ones(20,dtype=int)*2, np.ones(20,dtype=int)*3))
df_concat['label'] = label
df_concat
```

```python
df_concat.rename()
```

```python
df_concat.to_csv('classification_data.csv')
```
